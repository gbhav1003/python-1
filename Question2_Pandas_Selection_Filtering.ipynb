```python
import pandas as pd

# 1. Create the DataFrame
data = {
    'EmployeeID': [201, 202, 203, 204, 205, 206, 207, 208, 209, 210],
    'Name': ['Sarah', 'Michael', 'Jennifer', 'David', 'Lisa', 'James', 'Emma', 'Robert', 'Sophia', 'William'],
    'Department': ['Engineering', 'Sales', 'Marketing', 'Engineering', 'HR', 'Sales', 'Engineering', 'Marketing', 'Sales', 'HR'],
    'Salary': [85000, 62000, 71000, 92000, 58000, 68000, 78000, 75000, 65000, 60000],
    'Performance Score': [4, 3, 5, 5, 4, 4, 3, 4, 5, 3],
    'Years of Service': [5, 3, 7, 8, 4, 6, 2, 5, 4, 3]
}

df = pd.DataFrame(data)


print("Specific Columns:\n", df[['Name', 'Department', 'Salary']])
```

```python
print("\nEmployee at row 5:\n", df.iloc[5])
```

```python
df_indexed = df.set_index('EmployeeID')
print("\nEmployeeID 208:\n", df_indexed.loc[208])

```

```python
print("\nEngineering Department:\n", df[df['Department'] == 'Engineering'])
```

```python
print("\nSalary > 70000:\n", df[df['Salary'] > 70000])
```

```python

print("\nHigh Performing Sales:\n", df[(df['Department'] == 'Sales') & (df['Performance Score'] >= 4)])
```

```python

print("\nTop 3 Highest Paid:\n", df.sort_values(by='Salary', ascending=False).head(3))
```

```python
print("\nSorted by Dept and Salary:\n", df.sort_values(by=['Department', 'Salary'], ascending=[True, False]))
```
