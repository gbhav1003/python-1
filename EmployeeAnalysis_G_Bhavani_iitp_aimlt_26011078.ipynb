```python
import sqlite3
import pandas as pd
import random
```

#Task 1: Create Database

```python
conn = sqlite3.connect('employees.db')
cursor = conn.cursor()
cursor.execute('DROP TABLE IF EXISTS employees')
cursor.execute('''
    CREATE TABLE employees (
        emp_id INTEGER PRIMARY KEY,
        name TEXT,
        department TEXT,
        salary REAL,
        years_experience INTEGER,
        performance_score REAL
    )
''')
```

```python
departments = ['Engineering', 'Sales', 'Marketing', 'HR', 'Finance']
records = [(i, f'Employee_{i}', random.choice(departments),
            round(random.uniform(50000, 150000), 2),
            random.randint(1, 15),
            round(random.uniform(1.0, 5.0), 1)) for i in range(1, 41)]

cursor.executemany('INSERT INTO employees VALUES (?,?,?,?,?,?)', records)
conn.commit()
```

#Task 2: SQL Queries

```python
q1_sql = """SELECT name, department, salary, performance_score FROM employees
            WHERE performance_score >= 4.0 AND years_experience >= 3
            ORDER BY performance_score DESC LIMIT 15"""
df_q1 = pd.read_sql_query(q1_sql, conn)
df_q1
```

```python
q2_sql = """SELECT * FROM employees WHERE salary BETWEEN 70000 AND 110000
            AND department IN ('Engineering', 'Sales')
            ORDER BY department, salary DESC"""
df_q2 = pd.read_sql_query(q2_sql, conn)
df_q2
```

```python
q3_sql = """SELECT department, COUNT(*) as emp_count, AVG(salary) as avg_salary
            FROM employees GROUP BY department ORDER BY avg_salary DESC"""
df_q3 = pd.read_sql_query(q3_sql, conn)
df_q3
```

#Task 3: Pandas Analysis

```python
df_all = pd.read_sql_query("SELECT * FROM employees", conn)
df_all
```

```python
avg_perf = df_all.groupby('department')['performance_score'].mean()
avg_perf
```

```python
pandas_q1 = df_all.loc[
    (df_all['performance_score'] >= 4.0) & (df_all['years_experience'] >= 3),
    ['name', 'department', 'salary', 'performance_score']
].sort_values(by='performance_score', ascending=False).head(15)
pandas_q1
```
