```python
import requests
import pandas as pd

# Updated API Endpoint with required fields
url = "https://restcountries.com/v3.1/all?fields=name,population,region,area"

try:
    response = requests.get(url)

    print("HTTP Status Code:", response.status_code)

    if response.status_code == 200:
        data = response.json()
        print("Total Number of Countries Retrieved:", len(data))

        country_list = []

        for country in data:
            common_name = country.get("name", {}).get("common", "N/A")
            population = country.get("population", 0)
            region = country.get("region", "N/A")
            area = country.get("area", 0)

            country_list.append({
                "common_name": common_name,
                "population": population,
                "region": region,
                "area": area
            })

        df = pd.DataFrame(country_list)

        print("\nFirst 10 Rows of Extracted Data:")
        print(df.head(10))

        # Task 3 Summary
        region_population = df.groupby("region")["population"].sum()
        highest_population_region = region_population.idxmax()
        highest_population_value = region_population.max()

        largest_area_country = df.loc[df["area"].idxmax(), "common_name"]
        largest_area_value = df["area"].max()

        print("\n===== SUMMARY =====")
        print("Region with Highest Total Population:", highest_population_region)
        print("Total Population:", highest_population_value)

        print("\nCountry with Largest Area:", largest_area_country)
        print("Area:", largest_area_value, "sq. km")

    else:
        print("Failed to fetch data from API. Please check the endpoint.")

except requests.exceptions.RequestException as e:
    print("An error occurred:", e)
```
